```python
# import dependencies
import os
import numpy as np
import pandas as pd
```

```python
# read all CSV files into a pandas dataframe in which the year is the key for the dataframe
file_path = '../../collegescorecard/data'
dfs = {}
for filename in os.listdir(file_path):
    if 'MERGED' in filename:
        year = filename[6:][:7]
        dfs[year] = pd.read_csv(os.path.join(file_path, filename), encoding='utf-8', low_memory=False)
        
```

```python
# read the datadictionary on the third sheet from excel file
datadict = pd.read_excel(os.path.join('../../collegescorecard', 'CollegeScorecardDataDictionary.xlsx'), sheet_name=3)
```

```python
# retrieve datadictionary information for categories of interests
earn_cols = datadict[datadict['dev-category']=='earnings']
schl_cols = datadict[datadict['dev-category']=='school']
stud_cols = datadict[datadict['dev-category']=='student']
deg_cols = datadict[(datadict['dev-category']=='academics') &  (datadict['VARIABLE NAME'].str.contains('PCIP'))]
```

```python
# review if the datasets have reported earnings or family income
for key in sorted(dfs.keys(), reverse=True):
    print(f'dataset: {key}  ===============')
    print(dfs[key]['MN_EARN_WNE_P10'].isna().value_counts())
    print(dfs[key]['MN_EARN_WNE_P9'].isna().value_counts())
    print(dfs[key]['MN_EARN_WNE_P8'].isna().value_counts())
    print(dfs[key]['MN_EARN_WNE_P7'].isna().value_counts())
    print(dfs[key]['MN_EARN_WNE_P6'].isna().value_counts())
    print(dfs[key]['INC_N'].isna().value_counts())
    print(dfs[key]['DEP_INC_N'].isna().value_counts())
    print(dfs[key]['IND_INC_N'].isna().value_counts())
    print(dfs[key]['DEP_INC_AVG'].isna().value_counts())
    print(dfs[key]['IND_INC_AVG'].isna().value_counts())
    print(dfs[key]['FAMINC'].isna().value_counts())
    print(dfs[key]['FAMINC_IND'].isna().value_counts())
    print('\n---------------------------------\n')
```
