```python
import os
import pandas as pd
```

```python
path = os.getcwd()
files = [file for file in os.listdir(path) if file.endswith('.csv') ]
files
```

```python
dfs = { filename[:-4]: pd.read_csv(os.path.join(path, filename), encoding='utf-8', low_memory=False) for filename in files }
dfs.keys()
    
```

```python
for key in dfs.keys():
    print(f'df is {key}')
    print(f'columns: {dfs[key].columns.to_list()}')
    print(f'shape  : {dfs[key].shape}')
    print(dfs[key].head(2))
    print('\n-------------------\n')
```
